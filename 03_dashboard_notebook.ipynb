```python
df = spark.table("raw_youtube.default.clean_youtube_recommendation_dataset")
df.show(5)
```

```python
## total views by category
from pyspark.sql.functions import sum

dash_views_by_category = (
    df.groupBy("category_id")
      .agg(sum("view_count").alias("total_views"))
)

dash_views_by_category.write.mode("overwrite").saveAsTable("raw_youtube.default.dash_views_by_category")
```

```python
## avg engagement_rate by category
from pyspark.sql.functions import avg

dash_engagement_by_category = (
    df.groupBy("category_id")
      .agg(avg("engagement_rate").alias("avg_engagement_rate"))
)

dash_engagement_by_category.write.mode("overwrite").saveAsTable("raw_youtube.default.dash_engagement_by_category")
```

```python
## count by popularity_level
dash_popularity = df.groupBy("popularity_level").count()

dash_popularity.write.mode("overwrite").saveAsTable("raw_youtube.default.dash_popularity")
```
